```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(30,10))

axes[0].set_title('Top10_dummy')
ax = data[['Popularity','Top10_dummy']].groupby('Top10_dummy').Popularity.hist(ax=axes[0], bins=20, alpha=0.5,legend=True)
axes[0].set_xlabel('Popularity')
axes[0].set_ylabel('Top10_dummy count')

axes[1].set_title('Top50_dummy')
ax = data[['Popularity','Top50_dummy']].groupby('Top50_dummy').Popularity.hist(ax=axes[1], bins=20, alpha=0.5,legend=True)
axes[1].set_xlabel('Popularity')
axes[1].set_ylabel('Top50_dummy count')



```

```python
## correlation of each emotion with popularity
#  'Celebrate',
#  'Desire',
#  'Explore',
#  'Fun',
#  'Hope',
#  'Love',
#  'Nostalgia',
    
    
fig, axes = plt.subplots(nrows=4, ncols=2,figsize=(20,10))

fig.tight_layout()

axes[0][0].set_title('Celebrate')
ax = data[['Popularity','Celebrate']].groupby('Celebrate').Popularity.hist(ax=axes[0][0], bins=20, alpha=0.5,legend=True)
axes[0][0].set_xlabel('Popularity')
axes[0][0].set_ylabel('Celebrate')

axes[0][1].set_title('Desire')
ax = data[['Popularity','Desire']].groupby('Desire').Popularity.hist(ax=axes[0][1], bins=20, alpha=0.5,legend=True)
axes[0][1].set_xlabel('Popularity')
axes[0][1].set_ylabel('Desire')

axes[1][0].set_title('Explore')
ax = data[['Popularity','Explore']].groupby('Explore').Popularity.hist(ax=axes[1][0], bins=20, alpha=0.5,legend=True)
axes[1][0].set_xlabel('Popularity')
axes[1][0].set_ylabel('Explore')

axes[1][1].set_title('Fun')
ax = data[['Popularity','Fun']].groupby('Fun').Popularity.hist(ax=axes[1][1], bins=20, alpha=0.5,legend=True)
axes[1][1].set_xlabel('Popularity')
axes[1][1].set_ylabel('Fun')

axes[2][0].set_title('Hope')
ax = data[['Popularity','Hope']].groupby('Hope').Popularity.hist(ax=axes[2][0], bins=20, alpha=0.5,legend=True)
axes[2][0].set_xlabel('Popularity')
axes[2][0].set_ylabel('Hope')

axes[2][1].set_title('Love')
ax = data[['Popularity','Love']].groupby('Love').Popularity.hist(ax=axes[2][1], bins=20, alpha=0.5,legend=True)
axes[2][1].set_xlabel('Popularity')
axes[2][1].set_ylabel('Love')


axes[3][0].set_title('Nostalgia')
ax = data[['Popularity','Nostalgia']].groupby('Nostalgia').Popularity.hist(ax=axes[3][0], bins=20, alpha=0.5,legend=True)
axes[3][0].set_xlabel('Popularity')
axes[3][0].set_ylabel('Nostalgia')
plt.subplots_adjust(top=2)

plt.show()




```
