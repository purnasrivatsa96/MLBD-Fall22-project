```python
genre_labels = [
    'bolero',
    'boy band',
    'country',
    'dance/electronic',
    'else',
    'funk',
    'hip hop',
    'house',
    'indie',
    'jazz',
    'k-pop',
    'latin',
    'metal',
    'opm',
    'pop',
    'r&b/soul',
    'rap',
    'reggae',
    'reggaeton',
    'rock',
    'trap'
]
```

```python
# remove the existing genre columns
df = df.drop(genre_labels, axis='columns')
```

```python
# add one hot encoded genre columns
df[ohe.categories_[0]] = transformed.toarray()
```

```python
df.shape
```

```python
# One hot encode album/single
ohe = OneHotEncoder()
transformed = ohe.fit_transform(df[['Album/Single']])

df[ohe.categories_[0]] = transformed.toarray()
```

```python
# remove the existing album/single column and n-a column that adds no information
df = df.drop(['n-a', 'Album/Single'], axis='columns')
```

```python
# drop columns that are represented by other ohe columns
duplicate_labels = ['Genre', 'Genre_new', 'Country', 'Explicit', 'LDA_Topic']
df = df.drop(duplicate_labels, axis='columns')
```

```python
# remove unclear norm2 columns
norm_labels = [
    'anger_norm2',
    'anticipation_norm2',
    'disgust_norm2',
    'fear_norm2',
    'joy_norm2',
    'sadness_norm2',
    'surprise_norm2',
    'trust_norm2',
    'negative_norm2',
    'positive_norm2',
]
```

```python
df = df.drop(norm_labels, axis='columns')
```

```python
# remove unrelated columns
unrelated_labels = [
    'Released_after_2017',
    'Cluster',
    'Uri'
]
df = df.drop(unrelated_labels, axis='columns')
```

```python
# remove categorical columns that cannot be used, such as album title and artist name
categorical_labels = ['Title', 'Artist', 'Album']
df = df.drop(categorical_labels, axis='columns')
```

```python
df.shape
```

```python
df.head()
```

```python
# some release data have an 'n-a' value -- we want to drop those rows
df = df.replace('n-a', np.nan)
```

```python
df['Release_date'].unique().tolist()
```

```python
# extract day month and year as features from Release_data
df['Release_date'] = pd.to_datetime(df['Release_date'])
df['year'] = df['Release_date'].dt.year
df['month'] = df['Release_date'].dt.month
df['day'] = df['Release_date'].dt.day
```

```python
# let's drop the rows without a day, month, and year
df = df.dropna(subset=['year', 'month', 'day'])

# and drop the Release_date column now that it's been broken down
df = df.drop(['Release_date'], axis='columns')
```

```python
df.shape
```

```python
df.head()
```

```python
# convert string-y columns to nums
df = df.replace('n\x07', np.nan)
```

```python
object_types_int = [
 'Track_number',
 'Tracks_in_album',
 'duration_ms',
 'time_signature'
]

object_types_float = [
 'Artist_followers',
 'danceability',
 'energy',
 'key',
 'loudness',
 'mode',
 'speechiness',
 'acoustics',
 'instrumentalness',
 'liveliness',
 'valence',
 'tempo'
]
```

```python
# There are 41 rows with no Artist followers, we will drop these as followers is an important metric
# There are also 14 rows with no duration or time signature, which we will also drop
df = df.dropna(subset=['Artist_followers', 'tempo', 'duration_ms'])
```

```python
for otype in object_types_float:
    df[otype] = df[otype].astype(float)
```

```python
for otype in object_types_int:
    try:
        df[otype] = df[otype].astype(int)
    except:
        print(otype)
```

```python
df.dtypes.unique()
```

```python
# creating imputed dataset
imputed_df = df

cols_with_na = [val for val in df.columns if df[val].isna().sum() > 0]

imputed_df[cols_with_na] = df[cols_with_na].fillna(df[cols_with_na].mean())
```

```python
imputed_df = imputed_df.dropna(subset=['Artist_followers'])
```

```python
imputed_df.to_csv(OUTPUT_PATH + 'imputed_df.csv',index=False)
```
